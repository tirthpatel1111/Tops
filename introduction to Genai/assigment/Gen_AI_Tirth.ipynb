# Batch image classification

```python
import requests
import io
from PIL import Image
import torch
import torch.nn.functional as F
from transformers import ViTImageProcessor, ViTForImageClassification
import matplotlib.pyplot as plt
import numpy as np
from PIL import UnidentifiedImageError
```

```python
image_urls = [
    # Cats
    "https://media.4-paws.org/e/a/f/4/eaf42f3a9320224640395cf6e9610ff69a3c353b/VIER%20PFOTEN_2016-07-08_011-4993x3455-1920x1329.jpg",
    "https://media.4-paws.org/d/2/5/f/d25ff020556e4b5eae747c55576f3b50886c0b90/cut%20cat%20serhio%2002-1813x1811-720x719.jpg",
    "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTj0Yrx17n1drkMYpW55ZQkkXegIynoji4hJA&s",

    # Dogs
    "https://cdn.shopify.com/s/files/1/0086/0795/7054/files/Golden-Retriever.jpg?v=1645179525",
    "https://www.nylabone.com/-/media/project/oneweb/nylabone/images/dog101/10-intelligent-dog-breeds/golden-retriever-tongue-out.jpg?h=430&w=710",
    "https://d1m34p9xkycorf.cloudfront.net/labrador-retriever-dog-isolated-on-transparent-background-ai-generated-png.png"
]
```

```python
def load_image(url):
    try:
        resp = requests.get(url, timeout=10)
        img = Image.open(io.BytesIO(resp.content)).convert("RGB")
        return img
    except UnidentifiedImageError:
        print(f"❌ Could not decode image: {url}")
    except Exception as e:
        print(f"❌ Error loading image: {url}\n{e}")
```

```python
images = [img for img in (load_image(url) for url in image_urls) if img is not None]
```

```python
model_name = "google/vit-base-patch16-224"
```

```python
processor = ViTImageProcessor.from_pretrained(model_name)
model = ViTForImageClassification.from_pretrained(model_name)
model.eval()
```

```python
inputs = processor(images=images, return_tensors="pt")
pixel_values = inputs["pixel_values"]
```

```python
with torch.no_grad():
    logits = model(pixel_values).logits
    probs = F.softmax(logits, dim=-1)
```

```python
id2label = model.config.id2label
```

```python
for i, p in enumerate(probs):
    print(f"\nImage {i+1} Top-5 predictions:")
    top5 = torch.topk(p, 5)
    for idx, score in zip(top5.indices, top5.values):
        print(f"- {id2label[idx.item()]}: {score.item():.4f}")
```

```python
plt.figure(figsize=(15, 18))

for i, p in enumerate(probs):
    top5 = torch.topk(p, 5)
    labels = [id2label[idx.item()] for idx in top5.indices]
    scores = top5.values.numpy()

    plt.subplot(3, 2, i + 1)
    plt.barh(labels, scores)
    plt.title(f"Image {i+1} Class Probabilities")
    plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```
